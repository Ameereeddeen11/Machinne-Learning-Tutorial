```python
# import submodule pyplot 
import matplotlib.pyplot as plt
import numpy as np

# drawing x, y lines to out diagram 
xpoints = np.array([0, 4])
ypoints = np.array([0, 50])

# drawing diagram
plt.plot(xpoints, ypoints) # setting x, y lines and points
# plot() is function to draw points in diagram. Function draws a line from point to point 
# plot() needs 2 parameters x-axis and y-axis. 
# x-axis is the horizontal axis 
# y-axis is the vertical axis
'''
Points are at:
    1 point => from x is 0 and from y is 4 
    2 point => from x is 0 and from y is 50
    and plot function connect them with linear line
'''
plt.show() # drawing diagram
```
