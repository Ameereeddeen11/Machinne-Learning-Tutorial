```python
import matplotlib.pyplot as plt 
import numpy as np 

# plotting without line 
xpoints = np.array([1, 6])
ypoints = np.array([5,15])
'''
Points are:
    1 point => from x is 1 and from y is 5 
    2 point => from x is 6 and from y is 15
    and plot function will leave it alone without connect them with any line
'''
plt.plot(xpoints, ypoints, "o")
plt.show()
```

```python
x = np.array([2, 3, 8, 15, 20])
y = np.array([3, 5, 6, 12, 16])

plt.plot(x, y)
'''
Points are at:
    1 point => from x is 2 and from y is 3 
    2 point => from x is 3 and from y is 5
    3 point => from x is 8 and from y is 6
    ect.
    plot function will connect all points with linear line
'''
plt.show()
```

```python
# Default point from x or y 

# if we give only y points it will set x point default value 0, 1, 2, 3, ect. depending on the length of the y-points
ypoint = np.array([4, 2, 9, 7, 15, 1, 8])

plt.plot(ypoint)
plt.show()
```

```python
# if we give only y points it will set y point default value 0, 1, 2, 3, ect. depending on the length of the x-points
xpoint = np.array([5, 1, 8, 9, 10, 48, 12, 5])

plt.plot(xpoint)
plt.show()
```
