```python
import matplotlib.pyplot as plt
# I have imported pyplot in which are most of the Matplotlib utilities
import numpy as np

# plot() is function used to draw points in diagram
xpoints = np.array([0,4])
ypoints = np.array([0,50])

plt.plot(xpoints, ypoints) # function draws a line from point to point  
plt.show() 

#to plot only points
plt.plot(xpoints, ypoints, 'o') #'o' is shortcut use for plot only points
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#multiple points
xpoint = np.array([1,2,3,4,5,8])
ypoint = np.array([7,5,9,6,9,1])

plt.plot(xpoint, ypoint)
plt.show()

#points
plt.plot(xpoint, ypoint, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([1,3,4,5])
ypoint = np.array([2,4,7,5])

#each point with a specified marker
plt.plot(xpoint, ypoint, marker = 'o')#spot 'o' I can write also * or something like that 
plt.show()
```
